Handling duplicates:

to demonstrate, we can double up our movies DataFrame by appending it to itself

```python
temp_df=movies_df.append(movies_df)
temp_df.shape
```

using the append() will return a copy without afecting the original DataFrame.We are capturing this copy in temp so we arent working with the real data. using the .shape we see that our DataFrames rows have doubled.

now dropping the duplicates:

```python
temp_df=temp_df.drop_duplicates()
temp_df.shape
```

just like append(),the drop_duplicates() method will also return a copy of the dataframe with the duplicate values removed.

pandas has the inplace keyword argument on many of its methods. Using inplace=True will modify the DataFrame object in place:

```python
temp_df.drop_duplicates(inplace=True)
```

now our temp_df will have the transformed data automatically

another argument for drop_duplicates() is keep,which has three options:


1.  first:(default) drops the duplicates except the first one
2.  last: drop duplicates except for last occurence
3.  drop all duplicates.





Since we didn't define the keep arugment in the previous example it was defaulted to first. This means that if two rows are the same pandas will drop the second row and keep the first row. Using last has the opposite effect: the first row is dropped.

keep, on the other hand, will drop all duplicates. If two rows are the same then both will be dropped.

```python
temp_df = movies_df.append(movies_df)  # make a new copy

temp_df.drop_duplicates(inplace=True, keep=False)

temp_df.shape
```
